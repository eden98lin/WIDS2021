```python
import sys
if sys.version_info[0] < 3:
    raise Exception("Python 3 not detected")
    
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.feature_extraction import DictVectorizer
```

```python
df = pd.read_csv("C:/Users/Eden/Desktop/personal_projects/widsdatathon2021/TrainingWiDS2021.csv")
```

```python
dic = pd.read_csv("C:/Users/Eden/Desktop/personal_projects/widsdatathon2021/DataDictionaryWiDS2021.csv")
```

```python
labs_name = dic.loc[dic["Category"].isin(["labs", "labs blood gas"])]["Variable Name"].tolist()
labs_name
```

```python
sample_count = df.shape[0]
sample_count
```

```python
temp = df[labs_name[0]]
temp = temp.dropna()
np.sqrt(temp.var())
```

```python
mins = []
maxs = []
means = []
sds = []
na_rates = []
for name in labs_name:
    temp = df[name]
    temp = temp.dropna()
    na_rate = (sample_count - len(temp)) / sample_count
    mins.append(temp.min())
    maxs.append(temp.max())
    means.append(temp.mean())
    sds.append(np.sqrt(temp.var()))
    na_rates.append(na_rate)
    print("column: ", name, " has min: ", temp.min(), ", max: ", temp.max(), ", mean: ", temp.mean(), ", sd: ", np.sqrt(temp.var()), ", na rate: ", na_rate)
```

```python
comb_lists = list(zip(labs_name, mins, maxs, means, sds, na_rates))
comb_lists
```

```python
temp_df = pd.DataFrame(comb_lists, columns = ["variable", "min", "max", "mean", "sd", "na_rate"])
temp_df
```

```python
temp_df.to_excel("stats.xlsx")
```

```python
df = df.dropna(0)
```

```python
df.shape
```

```python
cat_names = df.columns.to_series().groupby(df.dtypes).groups.get(np.dtype('O')).tolist()
cat_names
```

```python
df_copy = df
```

```python
for i in range(len(cat_names)):
    # Code below adapted from:
    # https://stackoverflow.com/questions/23208745/adding-dummy-columns-to-the-original-dataframe
    df_copy = pd.concat([df_copy, pd.get_dummies(df_copy[cat_names[i]])], axis=1)
    df_copy = df_copy.drop(columns=[cat_names[i]])
```
